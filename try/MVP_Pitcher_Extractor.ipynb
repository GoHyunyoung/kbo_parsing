```python
import pandas as pd
import numpy as np
import urllib

from BoxScoreParser import BoxScoreParser
from UrlParser import UrlParser
from bs4 import BeautifulSoup
from sklearn import tree
from sklearn.externals.six import StringIO

class pitcher :
    def __init__(self):
        self.averageFault = 0
        self.numBatter = 0
        self.PA = 0
        self.Sout = 0
        self.U = 0
    def point(self):
        if not(self.averageFault) and not(self.numBatter) :
            return 0
        elif not(self.averageFault):
            return 0.5 + self.Sout/9 - self.PA/(self.numBatter*2)
        else :
            return (1/self.averageFault + (self.Sout+self.numBatter)/27 - self.PA/(self.numBatter*2))

date=20160522
urlParser = UrlParser(date)

nPitcher = pitcher()
mPitcher = pitcher()
atmp=[]
htmp=[]

i = len(urlParser.urlList)
indexnP = 0
indexmP = 0
for url in urlParser.urlList:
    print "ATeam", ;print i
    bxsParser = BoxScoreParser(url)
    
    import copy
    
    for j in range(len(bxsParser.a_pitRecord[u'평균자책점'])):
        nPitcher.averageFault = float(bxsParser.a_pitRecord[u'평균자책점'][j])
        nPitcher.numBatter = float(bxsParser.a_pitRecord[u'타자'][j])
        nPitcher.PA = float(bxsParser.a_pitRecord[u'피안타'][j])
        nPitcher.Sout = float(bxsParser.a_pitRecord[u'삼진'][j])
        atmp.append(copy.copy(nPitcher))
        
    for n in range(len(bxsParser.a_pitRecord[u'평균자책점'])):
        print atmp[n + indexnP].point(),
        print atmp[n + indexnP].averageFault,
        print atmp[n + indexnP].numBatter,
        print atmp[n + indexnP].PA,
        print atmp[n + indexnP].Sout,
        if atmp[n + indexnP].point() >= 1 :
            atmp[n + indexnP].U = 1
        print atmp[n + indexnP].U
    indexnP = indexnP + len(bxsParser.a_pitRecord[u'평균자책점'])
        
    print "HTeam", ;print i
    for k in range(len(bxsParser.h_pitRecord[u'평균자책점'])):
        mPitcher.averageFault = float(bxsParser.h_pitRecord[u'평균자책점'][k])
        mPitcher.numBatter = float(bxsParser.h_pitRecord[u'타자'][k])
        mPitcher.PA = float(bxsParser.h_pitRecord[u'피안타'][k])
        mPitcher.Sout = float(bxsParser.h_pitRecord[u'삼진'][k])
        htmp.append(copy.copy(mPitcher))
        
        
    for m in range(len(bxsParser.h_pitRecord[u'평균자책점'])):
        print htmp[m + indexmP].point(),
        print htmp[m + indexmP].averageFault,
        print htmp[m + indexmP].numBatter,
        print htmp[m + indexmP].PA,
        print htmp[m + indexmP].Sout,
        if htmp[m + indexmP].point() >= 1 :
            htmp[m + indexmP].U = 1
        print htmp[m + indexmP].U
    indexmP = indexmP + len(bxsParser.h_pitRecord[u'평균자책점'])
    i = i-1
```
