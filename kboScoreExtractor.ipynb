```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('http://www.koreabaseball.com/Schedule/ScoreBoard/ScoreBoard.aspx')
browser.find_element_by_css_selector('img.ui-datepicker-trigger').click()
time.sleep(0.01)

#http://www.koreabaseball.com/Schedule/ScoreBoard/ScoreBoard.aspx
#에서 몇달이나 뒤로가야 찾고자하는 달이 나오는지 range()에 넣기
#ex 현재 2016년 4월 찾고자하는달 : 2015년 9월 -> range(7)
for x in range(25):
    browser.find_element_by_css_selector('[title=이전달]').click()
    time.sleep(0.01)
    
l=list()
browser.find_elements_by_css_selector('[data-handler=selectDay]')[0].click()
for d in range(30):    
#30일의 경우 -> range(30)
#31일의 경우 -> range(31)
    time.sleep(0.1)
    for x in range(len(browser.find_elements_by_css_selector('table.tScore'))):
        time.sleep(0.1)
        try:
            tmp = browser.find_elements_by_css_selector('table.tScore tbody')[x].text
            l.append(tmp.strip().split())
        except:
            print 'Except occur!'
            time.sleep(1)
    browser.find_elements_by_css_selector('input#cphContainer_cphContents_btnNextDate')[0].click()
    #print browser.find_elements_by_css_selector('span#cphContainer_cphContents_lblGameDate')[0]
```

```python
for x in l:
    for y in x:
        print y+',',
    print
```

```python
f=file('test.csv','w+')
#해당월에 맞는 파일명 지정
for x in l:
    for y in x:
        f.write(y.encode('utf-8')+',')
    f.write('\n')
f.close()
```
