```python
#in:score[24]
def contextCheck(score):
    #print score
    #sumH = 홈팀점수합, sumA = 어웨이팀점수합
    sumH = 0
    sumA = 0
    for i in range(len(score)):
        if(i<12):
            sumH += score[i]
        else:
            sumA += score[i]
            
    #print sumA
    #print sumH
    #dist = 판별변수(점수가 차이나는 횟수), sH = 현재홈팀점수, sA = 현재어웨이팀점수, winner = 현재 리드중인 팀
    dist = 0
    sH = 0
    sA = 0
    winner = "?"
    if(sumH == sumA):
        #print "무승부(5)\n"
        #score[24] = 5
        return 5

    elif(( sumH == 0 or sumA == 0 )and ( sumH >= sumA + 6 or sumH + 6 <= sumA )):
        #print "영봉승(6)\n"
        #score[24] = 6
        return 6

    elif((sumH <= 3 and sumA >=9) or (sumH >= 9 and sumA <=3) or (sumH >= 7 and sumA >= sumH*2) or (sumH >= sumA*2 and sumA >= 7)):
        return 4
        #print "완승(4)\n"
        #score[24] = 4

    else:
        for j in range(len(score)/2):
            sH += score[j]
            sA += score[(j+12)]
            if(j<9 and (not(sH == 0 and sA ==0)) and (sH != sA)):
                if(sH > sA):
                    if(winner != "sH"):
                        dist += 1
                    winner = "sH"
                else:
                    if(winner != "sA"):
                        dist += 1
                    winner = "sA"

            elif((j>=9) and (dist<=1)):
                return 1
                #print "승리(1)\n"
                #print dist
                #score[24] = 1
                #break

            else:
                if(dist == 2):
                    return 3
                    #print "역전(3)"
                    #print dist
                    #score[24] = 3
                    #break
                elif(j>=9 and (dist>=3)):
                    return 2
                    #print "접전(2)"
                    #print dist
                    #score[24] = 2
                    #break
                #이후 데이터 생략
```

```python
for x in range(len(df[u'context'])):
    df[u'context'][x]=contextCheck(list(df.ix[x][:24]))
```
